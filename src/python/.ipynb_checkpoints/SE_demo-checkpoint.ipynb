## A Single Class with Seemingly Two Clusters


```python
# generate two clusters: a with 20 points, b with 20 points
np.random.seed(4711)    # repeatable experiment

labeled = 10
unlabeled = 100
samples = 2*(labeled+unlabeled)

mean1 = [10,0]
mean2 = [0, 20]
cov = [[3, 1], [1, 3]]

# cluster 1
X1 = np.random.multivariate_normal(mean1, cov, size=labeled)
U1 = np.random.multivariate_normal(mean1, cov, size=unlabeled)
Y1 = np.ones(labeled)
YU1 = np.zeros(unlabeled)

# cluster 2
X2 = np.random.multivariate_normal(mean2, cov, size=labeled)
U2 = np.random.multivariate_normal(mean2, cov, size=unlabeled)
Y2 = np.ones(labeled)
YU2 = np.zeros(unlabeled)

X = np.concatenate((X1, X2),)
U = np.concatenate((U1, U2),)


fig, ax = plt.subplots(figsize=(10,10))

ax.scatter(U[:,0], U[:, 1],
           marker='o', color='b', s=30,
           label='Unlabeled Points')
ax.scatter(X[:, 0], X[:, 1], 
           marker='o', color='r', s=30,
           label='Labeled Points')
ax.set_title('Given Dataset')
plt.legend()
plt.show()
```

```python
X = np.concatenate((X1, U1, X2, U2),)
Y = np.concatenate((Y1,YU1,Y2, YU2))
print('{n} samples with 2 dimensions:{shape}'.format(n=samples, shape=X.shape))
```

### Typical Laplacian Eigenmaps Contruction

```python
# find the knn neighbors
from utils.graph_construction import compute_adjacency, create_laplacian

W = compute_adjacency(X, n_neighbors=3)

#find the laplacian matrix for the original data and similarity matrix
D = np.diag(np.sum(W.toarray(),axis=1))
L = D-W.toarray()

# plot the matrix
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,7))

ax[0].spy(W, precision=1E-12, markersize=.75)
ax[1].spy(L, precision=1E-12, markersize=.75)
ax[0].set_title('Adjacency Matrix', fontsize=20)
ax[1].set_title('Laplacian Matrix', fontsize=20)
ax[0].tick_params(labelbottom='on', labeltop='off')
ax[1].tick_params(labelbottom='on', labeltop='off')
plt.show()
```

### Label Similarity Potential Matrix


```python
from manifold_learning.se import sim_potential
# new method for constructing the matrix
Ws1 = np.repeat(Y[:, np.newaxis], Y[:,].shape[0], axis=1)
Ws2 = np.repeat(Y[:, np.newaxis], Y[:,].shape[0], axis=1).T
Ws = np.zeros(np.shape(Ws1), dtype='int')
indices = np.where((Ws1==Ws2) & (Ws1>0) & (Ws2>0))
Ws[indices] = 1

#find the laplacian matrix for the original data and similarity matrix
Ds = np.diag(np.sum(Ws, axis=1))
V = Ds-Ws

# plot the matrix
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,7))

ax[0].spy(Ws, precision=1E-12, markersize=.75)
ax[1].spy(V, precision=1E-12, markersize=.75)
ax[0].set_title('Similarity Adjacency Matrix', fontsize=20)
ax[1].set_title('Potential Matrix', fontsize=20)
ax[0].tick_params(labelbottom='on', labeltop='off')
ax[1].tick_params(labelbottom='on', labeltop='off')
plt.show()
```

### Graph Representation

```python
# solve the eigenvalue problem
from scipy.linalg import eigh
alpha = 0.8
alpha = alpha * np.trace(L)/np.trace(V)

eigValsS, eigVecsS = eigh(a=L+alpha*V, b=D, 
                        eigvals=(0,L.shape[0]-1))
eigValsL, eigVecsL = eigh(a=L, b=D, 
                        eigvals=(0,L.shape[0]-1))

# plot the eigenvalues
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,7))

ax[0].plot(eigValsL, linewidth=2, label='Laplacian')
ax[1].plot(eigValsS, linewidth=2, label='Schroedinger')
ax[0].legend(loc='upper left')
ax[1].legend(loc='upper left')
plt.suptitle('Eigenvalues', fontsize=20)
plt.show()


# plot the embedding
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,7))

ax[1].scatter(eigVecsS[0, 1:3], eigVecsS[1, 1:3],
              marker='o',
              c='b', s=10, 
              label='Schroedinger')
ax[0].scatter(eigVecsL[0, 1:3], eigVecsL[1, 1:3],
              marker='o',
              c='b', s=10, 
              label='Laplacian')
ax[0].legend()

ax[1].legend()
plt.suptitle('Eigenvectors', fontsize=20)
plt.show()
```

So, this demonstration shows what happens when you add in the potential matrix with some labeled information to a cluster of points. You find that the cluster in the new embedding is much closer together. As you increase that $\alpha$ value, you get a tigher cluster. This is equivalent to making the eigenvalues go to 0.

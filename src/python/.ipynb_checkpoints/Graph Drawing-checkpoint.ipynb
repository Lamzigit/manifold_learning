```python
import networkx as nx
```

```python
# create an empty graph with no nodes or edges
G = nx.Graph()

# add 6 nodes
G.add_nodes_from([1, 2, 3, 4, 5, 6])

# add some edges
G.add_edges_from([(1,2), (1,3), (3,4), (2,5),
                  (1,6), (2,6)])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

nx.draw(G)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx

G=nx.Graph()

# add edges with weights
G.add_node('a')
G.add_node('b')
G.add_node('c')
G.add_node('d')
G.add_node('e')
G.add_node('f')
# position the nodes by Force layout
pos = nx.spring_layout(G)
nx.draw(G, pos)
nx.draw_networkx_nodes(G, pos)
plt.axis('off')
plt.savefig('figures/graph_noedges', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx

G=nx.Graph()

# add edges with weights
G.add_edge('a','b',weight=0.6)
G.add_edge('a','c',weight=0.2)
G.add_edge('c','d',weight=0.1)
G.add_edge('c','e',weight=0.7)
G.add_edge('c','f',weight=0.9)
G.add_edge('a','d',weight=0.3)

# position the nodes by Force layout
pos = nx.spring_layout(G)

# show the graph
nx.draw(G, pos)
plt.savefig('figures/graph_unweighted', dpi=300)
plt.show()

# add the edge weights
edge_weight=dict([((u,v,), float(d['weight'])) for u,v,d in G.edges(data=True)])

nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_weight)
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
plt.axis('off')
plt.savefig('figures/graph_weighted', dpi=300)
plt.show()
```

```python
A = nx.adjacency_matrix(G)
print (A.todense())

L = nx.laplacian_matrix(G)
print (L.todense())

spec = nx.laplacian_spectrum(G)
print spec
```

# Demo to Show what Semisupervised Manifold Alignment Does

```python
%matplotlib inline
from __future__ import print_function
from __future__ import division

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

So, this demonstration shows what happens when you add in the potential matrix with some labeled information to a cluster of points. You find that the cluster in the new embedding is much closer together. As you increase that $\alpha$ value, you get a tigher cluster. This is equivalent to making the eigenvalues go to 0.

## Multiple Classes in Multiple Clusters


```python
# generate two clusters: a with 20 points, b with 20 points
np.random.seed(47)    # repeatable experiment

labeled = 10
unlabeled = 100
samples = 2*(labeled+unlabeled)


# cluster I attributes

mean1 = [10,0]
mean2 = [8, 5]
cov = [[2.5, .9], [.9, 2.5]]

# labeled points
X11 = np.random.multivariate_normal(mean1, cov, size=labeled)
X12 = np.random.multivariate_normal(mean2, cov, size=labeled)
Y11 = np.ones(labeled)
Y12 = 2*np.ones(labeled)


# unlabeled points
U11 = np.random.multivariate_normal(mean1, cov, size=unlabeled)
U12 = np.random.multivariate_normal(mean2, cov, size=unlabeled)
YU11 = np.zeros(unlabeled)
YU12 = np.zeros(unlabeled)


# plot the first cluster
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(15,7))

# cluster I
ax[0].scatter(U11[:,0], U11[:, 1], 
              marker='o', c='k', s=30, 
              label='Unlabeled - Domain I')
ax[0].scatter(U12[:,0], U12[:, 1], 
              marker='o', c='m', s=30, 
              label='Unlabeled - Domain II')
ax[0].scatter(X11[:,0], X11[:, 1],
              marker='o', c='r', s=30, 
              label='Labeled - Domain I')
ax[0].scatter(X12[:,0], X12[:,1], 
              marker='o', c='b', s=30, 
              label='Labeled - Domain II')
ax[0].legend()

# cluster II

ax[1].scatter(U11[:,0], U11[:, 1], 
              marker='o', c='k', s=30, 
              label='Unlabeled - Domain I')
ax[1].scatter(U12[:,0], U12[:, 1], 
              marker='o', c='k', s=30, 
              label='Unlabeled - Domain II')
ax[1].scatter(X11[:,0], X11[:, 1],
              marker='o', c='r', s=30, 
              label='Labeled - Domain I')
ax[1].scatter(X12[:,0], X12[:,1], 
              marker='o', c='b', s=30, 
              label='Labeled - Domain II')
ax[1].legend()
plt.suptitle('Cluster I', size=20)
plt.show()

```

```python
# cluster II attributes

mean1 = [20,17]
mean2 = [25, 20]
cov = [[2.5, .9], [.9, 2.5]]

# labeled points
X21 = np.random.multivariate_normal(mean1, cov, size=labeled)
X22 = np.random.multivariate_normal(mean2, cov, size=labeled)
Y21 = np.ones(labeled)
Y22 = 2*np.ones(labeled)


# unlabeled points
U21 = np.random.multivariate_normal(mean1, cov, size=unlabeled)
U22 = np.random.multivariate_normal(mean2, cov, size=unlabeled)
YU21 = np.zeros(unlabeled)
YU22 = np.zeros(unlabeled)

# plot the second cluster
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(15,7))

# Labeled Unlabeled Attributes highlighted
ax[0].scatter(U21[:,0], U21[:, 1], 
              marker='o', c='k', s=30, 
              label='Unlabeled - Domain I')
ax[0].scatter(U22[:,0], U22[:, 1], 
              marker='o', c='m', s=30, 
              label='Unlabeled - Domain II')
ax[0].scatter(X21[:,0], X21[:, 1],
              marker='o', c='r', s=30, 
              label='Labeled - Domain I')
ax[0].scatter(X22[:,0], X22[:,1], 
              marker='o', c='b', s=30, 
              label='Labeled - Domain II')
ax[0].legend()


# Unlabeled Attributes
ax[1].scatter(U21[:,0], U21[:, 1], 
              marker='o', c='k', s=30, 
              label='Unlabeled - Domain I')
ax[1].scatter(U22[:,0], U22[:, 1], 
              marker='o', c='k', s=30, 
              label='Unlabeled - Domain II')
ax[1].scatter(X21[:,0], X21[:, 1],
              marker='o', c='r', s=30, 
              label='Labeled - Domain I')
ax[1].scatter(X22[:,0], X22[:,1], 
              marker='o', c='b', s=30, 
              label='Labeled - Domain II')
ax[1].legend()
plt.suptitle('Cluster II', fontsize=20)
plt.show()
```

```python
# concatentate them by class
X1 = np.concatenate((X11, X21),)
U1 = np.concatenate((U11, U21),)
Y1 = np.concatenate((Y11, Y21, YU11, YU21))

X2 = np.concatenate((X12, X22),)
U2 = np.concatenate((U12, U22),)
Y2 = np.concatenate((Y12, Y22, YU12, YU22))

# plot them
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,7))

ax[0].scatter(U1[:,0], U1[:,1],
              marker='o', color='k', s=10,
              label='Unlabeled - Domain I')
ax[0].scatter(U2[:,0], U2[:,1],
              marker='o', color='m', s=10,
              label='Unlabeled - Domain II')
ax[0].scatter(X1[:,0], X1[:,1], 
              marker='o', color='r', s=30, 
              label='Labeled - Domain I')
ax[0].scatter(X2[:,0], X2[:,1],
              marker='o', color='b', s=30,
              label='Labeled - Domain II')

ax[0].legend(loc='lower right')

ax[1].scatter(U1[:,0], U1[:,1],
              marker='o', color='k', s=10,
              label='Unlabeled - Domain I')
ax[1].scatter(U2[:,0], U2[:,1],
              marker='o', color='k', s=10,
              label='Unlabeled - Domain II')
ax[1].scatter(X1[:,0], X1[:,1], 
              marker='o', color='r', s=30, 
              label='Labeled - Domain I')
ax[1].scatter(X2[:,0], X2[:,1],
              marker='o', color='b', s=30,
              label='Labeled - Domain II')

ax[1].legend(loc='lower right')

plt.suptitle('Data Clusters', fontsize=20)
plt.show()
```

As you can see, I have successfully create two clusters of points that seem to be independent on each other. However, we find that the actual labels of these clusters seem to match the other cluster. This is an interesting scenerio because the obvious cut we can do to separate the data would be to draw a line between the two obvious clusters. But user input can force certain clusters to be together in the embedding which will give us a better cut.

### Individual Gemetric Matrices

```python
X1 = np.concatenate((X1,U1))
X2 = np.concatenate((X2,U2))
# find the knn neighbors
from utils.graph_construction import compute_adjacency

W1 = compute_adjacency(X1, n_neighbors=5)
W2 = compute_adjacency(X2, n_neighbors=5)

from scipy.linalg import block_diag

W = block_diag(W1.toarray(),W2.toarray())

#find the laplacian matrix for the original data and similarity matrix
Dg = np.diag(np.sum(W,axis=1))
Lg = Dg-W

# plot the matrix
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax[0].spy(W, precision=1E-12, markersize=.75)
ax[1].spy(Lg, precision=1E-12, markersize=.75)
ax[0].set_title('Adjacency Matrix')
ax[1].set_title('Laplacian Matrix')
ax[0].tick_params(labelbottom='on', labeltop='off')
ax[1].tick_params(labelbottom='on', labeltop='off')
plt.show()
```

### Similarity Potential Matrix

```python
Y = np.concatenate((Y1,Y2),)

from manifold_learning.se import sim_potential
# new method for constructing the matrix
Ws1 = np.repeat(Y[:, np.newaxis], Y[:,].shape[0], axis=1)
Ws2 = np.repeat(Y[:, np.newaxis], Y[:,].shape[0], axis=1).T
Ws = np.zeros(np.shape(Ws1), dtype='int')
indices = np.where((Ws1==Ws2) & (Ws1>0) & (Ws2>0))
Ws[indices] = 1

#find the laplacian matrix for the original data and similarity matrix
Ds = np.diag(np.sum(Ws, axis=1))
Vs = Ds-Ws

# plot the matrix
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax[0].spy(Ws, precision=1E-12, markersize=.75)
ax[1].spy(Vs, precision=1E-12, markersize=.75)
ax[0].set_title('Similarity Adjacency Matrix')
ax[1].set_title('Potential Matrix')
ax[0].tick_params(labelbottom='on', labeltop='off')
ax[1].tick_params(labelbottom='on', labeltop='off')
plt.show()
```

### Dissimilarity Potential Matrix

```python
from manifold_learning.se import sim_potential
# new method for constructing the matrix
Wd = np.zeros(np.shape(Ws1), dtype='int')
indices = np.where((Ws1!=Ws2) & (Ws1>0) & (Ws2>0))
Wd[indices] = 1

#find the laplacian matrix for the original data and similarity matrix
Dd = np.diag(np.sum(Wd, axis=1))
Vd = Dd-Wd

# plot the matrix
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax[0].spy(Wd, precision=1E-12, markersize=.75)
ax[1].spy(Vd, precision=1E-12, markersize=.75)
ax[0].set_title('Similarity Adjacency Matrix')
ax[1].set_title('Potential Matrix')
ax[0].tick_params(labelbottom='on', labeltop='off')
ax[1].tick_params(labelbottom='on', labeltop='off')
plt.show()
```

```python
# solve the eigenvalue problem
from scipy.linalg import eigh
alpha = 0.8
alpha = alpha * np.trace(Lg)/np.trace(Vs)

eigValsW, eigVecsW = eigh(a=Lg+alpha*Vs, b=Dg, 
                        eigvals=(0,Lg.shape[0]-1))

eigValsS, eigVecsS = eigh(a=Lg+alpha*Vs, b=Vd, 
                        eigvals=(0,Lg.shape[0]-1))


# plot the eigenvalues
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax[0].plot(eigValsW, label='Laplacian')
ax[1].plot(eigValsS, label='Schroedinger')
ax[0].legend(loc='upper left')
ax[1].legend(loc='upper left')
plt.suptitle('Eigenvalues', fontsize=12)
plt.show()


# plot the embedding
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax[1].scatter(eigVecsS[0, 1:3], eigVecsS[1, 1:3],
              marker='o',
              c='b', s=10, 
              label='Schroedinger')
ax[0].scatter(eigVecsW[0, 1:3], eigVecsL[1, 1:3],
              marker='o',
              c='b', s=10, 
              label='Laplacian')
ax[0].legend()

ax[1].legend()
plt.suptitle('Eigenvectors', fontsize=12)
plt.show()
```
